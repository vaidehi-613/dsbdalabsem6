```python
print(quantile1,quantile3)
```

```python
iqr_value = (quantile3 - quantile1)
```

```python
iqr_value
```

```python
lb_value = quantile1 - (1.5*iqr_value)
```

```python
ub_value = quantile3 + (1.5*iqr_value)
```
